```python
#Time series analysis
import pandas as pd
df = pd.read_csv('aapl.csv',parse_dates = ['Date'],index_col = 'Date') #To change  the index as date we used index_col
df
```

```python
#1)Datetimeindex
df.index
```

```python
#To get data of only jan 2017 
df.loc['2017-01']
```

```python
df.loc['2017-01'].Close.mean()
```

```python
#2)Resampling
#Suppose I m not interested in daily stock prices of apple but i want to know montly stock price we can use resampling
df.Close.resample('M').mean()
```

```python
%matplotlib inline
df.Close.resample('M').mean().plot()
```

```python
%matplotlib inline
df.Close.resample('Q').mean().plot(kind = 'bar')
```
